```python
#System Modules
import os
import sys
sys.path.append(os.path.abspath(os.path.join('..')))
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import streamlit as st
import matplotlib.pyplot as plt

import script.ploting_fun as plot
# User Averview Analysis
st.set_option('deprecation.showPyplotGlobalUse', False)
def get_overview():
    top_10_handset = df_overview.groupby("Handset Type")['MSISDN/Number'].nunique().nlargest(10)
    top_3_manufacturers = df_overview.groupby("Handset Manufacturer")['MSISDN/Number'].nunique().nlargest(3)

    fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,8))
    plot.serious_bar(top_3_manufacturers, ax1)
    
    plot.serious_bar(top_10_handset, ax2)
    plt.xticks(rotation=65)
```
